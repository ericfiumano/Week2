```python
import pandas as pd

# Load the datasets
customer_data = pd.read_csv('Customer_ID.csv')
cab_data = pd.read_csv('Cab_Data.csv')
transaction_data = pd.read_csv('Transaction_ID.csv')
city_data = pd.read_csv('City.csv')
weather_data = pd.read_csv('WeatherEvents_Jan2016-Dec2022.csv')
holiday_data = pd.read_csv('US Holiday Dates (2004-2021).csv')

print("Customer Data:")
print(customer_data.info())

print("\nCab Data:")
print(cab_data.info())

print("\nTransaction Data:")
print(transaction_data.info())

print("\nCity Data:")
print(city_data.info())

print("\nWeather Data:")
print(weather_data.info())

print("\nHoliday Data:")
print(holiday_data.info())




# Merge relevant datasets
merged_data = pd.merge(cab_data, transaction_data, on='Transaction ID')
merged_data = pd.merge(merged_data, customer_data, on='Customer ID')
merged_data = pd.merge(merged_data, city_data, on='City')

# Group by Cab Type and perform forecasting
cab_profit_forecast = merged_data.groupby('Company')[['Price Charged', 'Cost of Trip']].sum()
cab_ride_forecast = merged_data.groupby('Company')['Transaction ID'].count()

# Display the forecasts
print("Profit Forecast for Each Cab Type:")
print(cab_profit_forecast)

print("\nNumber of Rides Forecast for Each Cab Type:")
print(cab_ride_forecast)

# Calculate profitability ratio for each cab company
cab_profitability = (cab_profit_forecast['Price Charged'] - cab_profit_forecast['Cost of Trip']) / cab_profit_forecast['Cost of Trip']

# Find the most profitable cab company
most_profitable_company = cab_profitability.idxmax()

print("The most profitable cab company is:", most_profitable_company)
```
